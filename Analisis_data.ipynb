## Hasil Assessing Data

Setelah dilakukan proses menilai kualitas data (assessing data) sebelum analisis, berikut adalah kesimpulan informasi yang diperoleh:

    MISSING VALUES   ---> 4 kolom memiliki nilai null, yaitu [children, country, agent, company].

    TIPE DATA        ---> Terdapat kesalahan pada tipe data di kolom reservation_status_date, yang awalnya berbentuk objek. Kolom ini perlu dikonversi menjadi tipe tanggal (date) agar dapat diproses lebih lanjut.

    DUPLICATE DATA   ---> Ditemukan 31.994 baris duplikat dalam data.

> Solusi untuk missing value

Kolom **children**, **country**, **agent**, dan **company** mengandung nilai missing (NaN). Terdapat tiga metode umum untuk menangani missing value, yaitu *dropping*, *imputation*, dan *interpolation*. Dalam kasus ini, metode ***dropping*** dipilih dengan alasan berikut:
- **Jumlah Data yang Hilang Sedikit**: Hanya sebagian kecil data yang memiliki nilai NaN, sehingga penghapusan tidak akan berdampak signifikan pada jumlah data yang tersedia atau analisis yang dilakukan.
- **Data yang Hilang Tidak Penting**: Baris dengan nilai hilang tidak mengandung informasi krusial, sehingga penghapusan lebih praktis dan menghindari potensi bias dari metode imputation.
- **Kolom Tidak Relevan**: Kolom **agent** dan **company** dianggap tidak relevan dengan variabel pembatalan pemesanan, sehingga kedua kolom ini akan dihapus dari dataset.

> Solusi untuk tipe data

- Kolom **reservation_status_date** akan dikonversi dari tipe objek ke tipe *datetime*.

> Solusi untuk duplikasi data

- Baris duplikat akan dihapus untuk menghilangkan redundansi data.

## - Jumlah pesanan dibatalkan berdasarkan tipe deposit

```python
cancellation_counts = df.groupby(['deposit_type', 'is_canceled']).size().unstack(fill_value=0)

# Reset index to make it easier to work with
cancellation_counts = cancellation_counts.reset_index()

# Melt the DataFrame to have a long format for seaborn
cancellation_counts_melted = cancellation_counts.melt(id_vars='deposit_type', 
                                                      value_vars=[0, 1], 
                                                      var_name='cancellation_status', 
                                                      value_name='count')

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=cancellation_counts_melted, 
            x='deposit_type', 
            y='count', 
            hue='cancellation_status', 
            palette='Set1')

plt.xlabel('Deposit Type')
plt.ylabel('Count')
plt.title('Jumlah pesanan daibatalkan berdasarkan tipe deposit')
plt.legend(title='Cancellation Status', labels=['Not Canceled (0)', 'Canceled (1)'])
plt.show()
```

Seperti yang grafik di atas, sebagian besar pesanan yang dibatalkan dilakukan dengan opsi "No deposit"

## - Perbandingan jumlah pesanan yang diterima dan dibatalkan antara City Hotel dan Resort Hotel

```python
# Set the background color of the figure
plt.figure(figsize=(8, 4), facecolor='#f7f7f7')  # Warna latar belakang yang cerah

# Create the countplot
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette="coolwarm")

# Customize legend location
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))

# Set plot title and axis labels
plt.title('Reservation Status in Different Hotels', size=20, color='#333333')  # Judul dengan warna abu gelap
plt.xlabel('Hotel', color='#333333', fontsize=12)
plt.ylabel('Number of Reservations', color='#333333', fontsize=12)

# Customize legend labels
plt.legend(['Not Cancelled', 'Cancelled'], loc='upper right', fontsize=10, title="Status")

# Annotate bars with data labels for exact counts
for p in ax1.patches:
    ax1.annotate(f'{int(p.get_height())}',
                 (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', xytext=(0, 10),
                 textcoords='offset points', fontsize=10, color='black')


# Calculate and print the counts per hotel for Cancelled/Not Cancelled
counts = df.groupby(['hotel', 'is_canceled']).size().unstack(fill_value=0)
print("\njumlah pesanan yang diterima dan dibatalkan antara City Hotel dan Resort Hotel:")
for hotel in counts.index:
    print(f"Hotel {hotel} - Not Cancelled: {counts.loc[hotel, 0]}, Cancelled: {counts.loc[hotel, 1]}")

# Show the plot
plt.show()
```

Pada hasil perbandingan tipe hotel di atas, didapat bahwa City Hotel memiliki lebih banyak pemesanan serta pembatalan pemesanan yang lebih tinggi, hal ini Mungkin karena Hotel Resort lebih mahal daripada Hotel City.

## - Tarif harian rata-rata di city hotel dan resort hotel

```python
resort_hotel= df[df['hotel']=='Resort Hotel']
City_Hotel = df[df['hotel']== 'City Hotel']

resort_hotel_rate = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
City_Hotel_rate = City_Hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8), facecolor='#C38154')
plt.title('Tarif Harian Rata-rata di city hotel dan resort hotel', fontsize=30)
plt.plot(resort_hotel_rate.index,resort_hotel_rate['adr'],label = 'Resort Hotel')
plt.plot(City_Hotel_rate.index,City_Hotel_rate['adr'],label = 'City Hotel')
plt.legend(fontsize=20)
plt.show()
```

Grafik garis di atas menunjukkan bahwa pada hari-hari tertentu, tarif harian rata-rata untuk City Hotel lebih rendah dibandingkan dengan tarif harian Resort Hotel. Selain itu, terdapat beberapa hari di mana tarif harian City Hotel bahkan lebih rendah lagi. Hal ini menunjukkan kemungkinan bahwa pada akhir pekan dan hari libur, tarif Resort Hotel cenderung mengalami kenaikan.

## - Analisis pendapatan rata-rata per hari (ADR) berdasarkan bulan.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat plot ADR per bulan untuk reservasi yang dibatalkan
plt.figure(figsize=(15, 8))
plt.title('Pendapatan ADR per bulan', fontsize=30)

# Data untuk ADR berdasarkan bulan dan status pembatalan
data = df[df['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index()

# Plot dengan warna hijau
sns.barplot(x='month', y='adr', data=data, color='green')

# Menambahkan legenda dan menampilkan plot
plt.legend(fontsize=20)
plt.show()
```

```python
cancaled_df_adr = cancaled_df_adr[(cancaled_df_adr['reservation_status_date']>'2016') & (cancaled_df_adr['reservation_status_date']<'2017-09')]
not_cancaled_df_adr = not_cancaled_df_adr[(not_cancaled_df_adr['reservation_status_date']>'2016') & (not_cancaled_df_adr['reservation_status_date']<'2017-09')]
```
